```python
'''
list of employees 

find hra then net salary 

iterate for each employee 
    then find hra from hra_per and then net salary 
'''
employees = [
    {'name': 'Jaiswal', 'age': 23, 'salary': 300000, 'is_active': True, 'hra_per': 20},
    {'name': 'Suryavansy', 'age': 14, 'salary': 150000, 'is_active': True, 'hra_per': 20},
    {'name': 'Sreyes', 'age': 28, 'salary': 350000, 'is_active': True, 'hra_per': 10}
]

for emp in employees:
    salary = emp['salary']
    hra_per = emp['hra_per']
    age = emp['age']
    if age < 21:
        continue
    #
    hra = (salary / 100) * hra_per
    net_salary = salary + hra 

    emp['hra'] = hra
    emp['net_salary'] = net_salary
# 

print(employees)
```
