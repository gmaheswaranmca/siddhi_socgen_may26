```python
# list of ages
ages = [10, 26, 2, 5, 18, 20, 13, 7, 8, 14]

print(ages)
```

```python
# 1. max age of first 2 ages
age1 = ages[0]
age2 = ages[1] 

if age1 > age2:
    max_age = age1
else:
    max_age = age2

# print('Max of', age1, 'and', age2, 'is', max)
'''
print('Max of', age1, 'and', age2, 'is', max, sep='\n', end='')
print('End of Program')
'''
print(f'Max of {age1} and {age2} is {max_age}.')

print('Max of %d and %d is %d.'%(age1, age2, max_age))
```

```python
# 2. max age of first 3 ages
age1 : int = ages[0]
age2 : int = ages[1]
age3 : int = ages[2]

if age1 > age2 and age1 > age3:
    max_age = age1
elif age2 > age3:
    max_age = age2
else:
    max_age = age3

print(f'Max of {age1}, {age2} and {age3} is {max_age}')
```

```python
# 2. max age of first 3 ages - version 2
age1 : int = ages[0]
age2 : int = ages[1]
age3 : int = ages[2]

max_age = age1
if age2 > max_age:
    max_age = age2
#
if age3 > max_age:
    max_age = age3
#

print(f'Max of {age1}, {age2} and {age3} is {max_age}')
```

```python
# print age from list using range/indexed loop 
for I in range(10): # 0 1 2 .. <10
    print(ages[I], end=' ')

print('\n')
# print age from list using element iterator loop 
for age in ages:
    print(age, end=', ')

print('\n')
I = 0
while I < 10:
    print(ages[I], end=' ')
    I += 1
```

```python
# 3. max from list 
max_age = ages[0]
for age in ages:
    if age > max_age:
        max_age = age

print(f'max {ages} is {max_age}')
```

```python
print(f'max {ages} is {max(ages)}')
```

```python
# avg of age
s = 0
for age in ages:
    s += age

avg = s // len(ages)

print(f'sum is {s}')
print(f'average is {avg}')
```

```python
# avg of age
s = sum(ages)
avg = s // len(ages)

print(f'sum is {s}')
print(f'average is {avg}')
```

```python
# 5. count age "above teen age" and "below or equal teen age"
above_teens = 0
others = 0

for age in ages:
    if age > 19:
        above_teens += 1
    else:
        others += 1
    #
#

print(f'ages are {ages}.')
print(f'above teenagers is #{above_teens}.')
print(f'below or equal teenagers is #{others}.') 
```

```python
# 6. Increase every age by 1
print(f'before increment : ages are {ages}.')
for I in range(10):
    ages[I] += 1
print(f'after increment : ages are {ages}.')
```

```python
# 7. Add one more age into list
age11 = 23
ages.append(age11)
print(f'after add: {ages}')
```

```python
# 8. Update 2nd age element as plus 2 
ages[1] += 2
print(f'after update: {ages}')
```
