```python
# emps list each emp is tuple
employees = []
employees.append(('Jaiswal',22))
employees.append(('Suryavansy',14))
employees.append(('Yadhav',28))
employees.append(('Archer',30))
print(employees)  

max_aged = employees[0]
for emp in employees:
    if emp[1] > max_aged[1]: # at index 0 == name attr, at index 1==age attr
        max_aged = emp
    #
#

print(f'Max aged: {max_aged}')
```

```python
# emps list each emp is dict
employees = []
employees.append({'name':'Jaiswal','age':22})
employees.append({'name':'Suryavansy','age':14})
employees.append({'name':'Sreyes','age':28})
employees.append({'name':'Archer','age':30})
print(employees)  

max_aged = employees[0]
for emp in employees:
    if emp['age'] > max_aged['age']: 
        max_aged = emp
    #
#

print(f'Max aged: {max_aged}')
```

```python
# emps list each emp is dict
# update every employee age by 1
employees = []
employees.append({'name':'Jaiswal','age':22})
employees.append({'name':'Suryavansy','age':14})
employees.append({'name':'Sreyes','age':28})
employees.append({'name':'Archer','age':30})
print(employees)  

max_aged = employees[0]
for emp in employees:
    emp['age'] += 1
#

print(employees) 
```

```python
import csv
with open('update_employees1.csv', 'w', newline='') as out:
    fieldnames = ['name', 'age']
    writer = csv.DictWriter(out, fieldnames=fieldnames)
    writer.writeheader()
    for emp in employees:
        writer.writerow(emp)
```
