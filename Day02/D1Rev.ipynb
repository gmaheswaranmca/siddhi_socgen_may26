```python
# Python Data -> Primitive, Composite-Derived
name = 'Mahesh'
age = 47
salary = 10000.75
is_active = True 

ages = [1,5,6,10,1,5,8,9,10,1] # list -> indexed collection -> sequence -> read / write
ages_tuple = (1,5,6,10,1,5,8,9,10,1) # tuple -> indexed collection -> sequence -> readonly
ages_freq = {1: 3, 5 : 2, 6: 1, 10:2, 8:1, 9:1} # dict -> key:value pair collection -> keys are no duplicates
    # key : primitive or tuple # hashing
ages_distict = {1, 5, 6, 10, 8, 9} # set -> no duplicate collection -> keys collection -> hashing 
    # order is not inserting order 

print(name, age, salary, is_active)
print(ages)
print(ages_tuple)
print(ages_freq)
print(ages_distict)

'''
to have list of composite data ie list of employees, list of flights
each composite data can be of tuple / list / dict / object of class
'''
```
