```python
# function 
# module # package

# what is function?
#    named code, with args for input, return value for output
#    reusability
#    task as a function -> task code

# pass by value
def disp_emp(name, age, salary):
    print(f'Name: {name}')
    print(f'Age: {age}')
    print(f'Salary: {salary}')
    print()
#

disp_emp('Jaiswal', 23, 300000) # positional arg
disp_emp('Suryavansy', 14, 150000)

disp_emp(name = 'Jaiswal',  age = 23, salary = 300000) # keyword arg
disp_emp(name = 'Jaiswal',  salary = 300000, age = 23) # keyword arg order changed
```

```python
# default arg -> function overloading -> redefining fn == same fn name with new code
def process_pay(salary, hra_per = 0):
    print(salary)
    print(hra_per)

process_pay(350000, 20)
process_pay(350000)
process_pay(350000, hra_per=20)
```

```python
# rutrun value out of function
# update the global variable 
net_salary = 0
def find_net_salary(salary, hra_per = 0, da_per = 10):
    print(salary)
    print(hra_per)
    print(da_per)
    hra = (hra_per / 100) * salary
    da = (da_per / 100) * salary
    global net_salary # to update the global varialbe, mark using global
    net_salary = salary + hra + da    
    return net_salary

print(find_net_salary(300000, 20, 10))
print(net_salary)
```

```python
# multiple return values
# returned as tuple
# in function call, expand / destructure the tuple to variables
def find_net_salary_v2(salary, hra_per = 0, da_per = 10):
    hra = (hra_per / 100) * salary
    da = (da_per / 100) * salary
    net_salary = salary + hra + da    
    return hra, da, net_salary

print(find_net_salary_v2(300000, 20, 10))
hra, da, net_salary = find_net_salary_v2(300000, 20, 10)
print(hra, da, net_salary)
```

```python
# variable args is defined using '*' prefixed name
def find_sum_of_n(*nums):
    print(type(nums))
    result = 0
    for num in nums:
        result += num
    return result

print(find_sum_of_n(10))
print(find_sum_of_n(10, 20))
print(find_sum_of_n(10, 20, 30))
print(find_sum_of_n(10, 20, 30, 40, 50, 60))
```

```python
# keyword variable args is defined using '**' prefixed name
def find_sum_of_n_ka(**nums):
    print(type(nums))
    result = 0
    for num in nums:
        result += nums[num]
    return result

print(find_sum_of_n_ka(num1=10))
print(find_sum_of_n_ka(num1=10, num2=20))
print(find_sum_of_n_ka(num1=10, num2=20, num3=30))
print(find_sum_of_n_ka(num1=10, num2=20, num3=30, num4=40, num5=50, num6=60))
```

```python
# function call within another function
def find_per_val(base, per):
    result = (per / 100) * base
    return result
    
def find_net_salary_v3(salary, hra_per = 0, da_per = 10):
    hra = find_per_val(salary, hra_per) # find_per_val(per=hra_per, base=salary)
    da = find_per_val(salary, da_per)
    net_salary = salary + hra + da    
    return net_salary

print(find_net_salary_v3(300000, 20, 10))
```

```python
# function defintion within another function

def find_net_salary_v4(salary, hra_per = 0, da_per = 10):
    def find_per_val(base, per):
        result = (per / 100) * base
        return result
        
    hra = find_per_val(salary, hra_per) # find_per_val(per=hra_per, base=salary)
    da = find_per_val(salary, da_per)
    net_salary = salary + hra + da    
    return net_salary

print(find_net_salary_v4(300000, 20, 10))
```

```python
# unamed function : without name
# see experession by fn
# for given arg, form expr that will be returned without return keyword
# use cases::higher order function map, filter -> map lambda (tranformation) or filter lambda(condition)
find_sum_of_2 = lambda num1, num2 : num1 + num2
print(find_sum_of_2(1,2))

print((lambda num1, num2 : num1 + num2)(3,4)) # lambda fn is called directly without name
```

```python
# lambda function defintion within another function

def find_net_salary_v5(salary, hra_per = 0, da_per = 10):
    # lambda fn def
    find_per_val = lambda base, per: (per / 100) * base
    # logic code    
    hra = find_per_val(salary, hra_per) # find_per_val(per=hra_per, base=salary)
    da = find_per_val(salary, da_per)
    net_salary = salary + hra + da    
    return net_salary

print(find_net_salary_v5(300000, 20, 10))
```
