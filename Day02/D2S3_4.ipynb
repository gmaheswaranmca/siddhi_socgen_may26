```python
# recursion or recursive function
# function calls itself is called recursive function 
# n! = n * (n-1)!
# 5! = 5 * 4!
# use cases: backtracking, dfs of tree traversal, etc

def fact(n):
    if n <= 1: # base/edge cond
        return 1
    #
    return n * fact(n-1) # recursive case

fact_of_5 = fact(5)
print(fact_of_5)
#       5 * fact(4)
#       5 * 4 * fact(3)
#       5 * 4 * 3 * fact(2)
#       5 * 4 * 3 * 2 * fact(1) <- call frame 5
#       5 * 4 * 3 * 2 * 1 <- cf5
#       5 * 4 * 3 * 2 <- cf4
#       5 * 4 * 6 <- cf3
#       5 * 24 <- cf2
#       120 <- cf1 ie mcf
```

```python
# var arg fn 

# find max of n nums
# define function 
# call by value
def find_max_of_n(num1, num2, num3 = None, *others):
    max_num = num1 
    if num2 > max_num: # max of 2
        max_num = num2
    #
    
    if num3 == None:
        return max_num 
    #

    if num3 > max_num: # max of 3
        max_num = num3
    #

    if len(others) == 0: 
        return max_num
    #

    for other_num in others: # max of n
        if other_num > max_num:
            max_num = other_num
        #
    #
    return max_num 
#

# positional args
print(find_max_of_n(10, 20))
print(find_max_of_n(10, 20, 30))
print(find_max_of_n(5.2, 7.4, 3.1))
print(find_max_of_n((8,10), (4,20), (2,50)))

# keyword args
print(find_max_of_n(num1=10, num2=20, num3=30)) 
print(find_max_of_n(num3=10, num1=20, num2=30)) 

# variable args 
print(find_max_of_n(10, 20, 40, 30))
print(find_max_of_n(10, 20, 40, 30, 70, 25))
```

```python
# var keyword arg fn 

# find max of n nums
# define function 
# call by value
def find_max_of_n(num1, num2, num3 = None, **others):
    max_num = num1 
    if num2 > max_num: # max of 2
        max_num = num2
    #
    
    if num3 == None:
        return max_num 
    #

    if num3 > max_num: # max of 3
        max_num = num3
    #

    if len(others) == 0: 
        return max_num
    #

    for other_key in others: # max of n
        other_num = others[other_key]
        if other_num > max_num:
            max_num = other_num
        #
    #
    return max_num 
#

# keyword variable args 
print(find_max_of_n(num1 = 10, num2 = 20, num3 = 40, num4 = 30))
print(find_max_of_n(num1 = 10, num2 = 20, num3 = 40, num4 = 30, num5 = 70, num6 = 25))
```
