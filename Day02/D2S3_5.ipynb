```python
# using std module 
import math
print(math.sqrt(25))
```

```python
# using user defined module 
import calc 
print(calc.find_sum(15,10))
```

```python
# alias name for module
import math as mymath
print(mymath.sqrt(25))
```

```python
# expand the name from module 
# importing the name (var/fn/cls etc) from module
from math import sqrt
print(sqrt(25))
```

```python
# alias name for name import from module 
from math import sqrt as square_root
print(square_root(25))
```
