```python
class InvalidDenominator(Exception):
    pass

try:
    num = int(input('Number:'))

    if num != 0:
        quotient = 1002 / num
    else:
        raise InvalidDenominator('I have to handle the denominator')

    print(quotient)
except ValueError as ex:
    print(ex)
except ZeroDivisionError as ex:
    print(ex)
except InvalidDenominator as ex: # custon exception 
    print(ex)
except Exception as ex: # generic exception 
    print(ex)
else:
    print('I completed my logic')
finally:
    print('Cleaning up - release resource')


print('End of Program')
```
