```python
### Map-comprehensions and filter-comprehensions
nums = [1, 2, 3, 4, 5, 6]

# map using comprehension
mapped_result = [num * 10 for num in nums]

# filter using comprehension
filtered_result = [num for num in nums if num % 2 == 0]

# map + filter together
map_filter_result = [x * 10 for x in nums if x % 2 == 0]


print("Original List")
print(nums)
print()

print("Map Comprehension")
print(mapped_result)
print()

print("Filter Comprehension")
print(filtered_result)
print()

print("Map + Filter Comprehension")
print(map_filter_result)
```

```python
### List, Tuple, Set and Dictionary comprehensions
# Comprehensions in Python
nums = [1, 2, 3, 4, 5, 5, 2]

# List Comprehension
list_result = [x * 2 for x in nums]

# Tuple Comprehension
tuple_result = tuple(x * 2 for x in nums)

# Set Comprehension
set_result = {x * 2 for x in nums}

# Dictionary Comprehension
dict_result = {x: x * 2 for x in nums}

print("List Comprehension")
print(list_result)
print()
print("Tuple Comprehension")
print(tuple_result)
print()
print("Set Comprehension")
print(set_result)
print()
print("Dictionary Comprehension")
print(dict_result)
```

```python
def fibonacci(n):
    first, second = 0,1
    for I in range(n):
        yield first
        third = first + second
        first, second = second, third

gen = fibonacci(10)
for term in gen:
    print(term)
```

```python
class Fib:
    def __init__(self, n):
        self.n = n
        self.first = 0
        self.second = 1
        self.I = 0
    def __iter__(self):
        return self
    def __next__(self):
        if (self.I + 1) >= self.n:
            raise StopIteration
        result = self.first
        
        third = self.first + self.second
        self.first, self.second = self.second, third
        self.I += 1

        return result 

fib_iter = Fib(5)
for num in fib_iter:
    print(num)
    
```

```python
class Account: 
    def __init__(self, name, number, amount):
        self.name = name
        self.number = number
        self.__balance = amount
    def __repr__(self):
        return f'[name={self.name}, number={self.number}, balance={self.__balance}]'
    def __str__(self):
        return f'<{self.name}, {self.number}, {self.__balance}>'
    def deposit(self, amount):
        self.__balance += amount
    def withdraw(self, amount):
        if self.__balance < amount:
            print('Issufficient balance')
            return 
        self.__balance -= amount

ac1 = Account('Jaiswal', '1001', 50000)
ac2 = Account('Suryavansy', '1002', 20000)
print(ac1, ac2)
accounts = [ac1, ac2]
print(accounts)

ac2.deposit(5000)
ac2.deposit(3000)
ac2.withdraw(1000)
print(accounts)
ac2.__balance = 0
print(accounts)
print(ac2.__dict__)
```

```python
# inheritance - Person -> Employee
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


    def show_details(self):
        print("Name :", self.name)
        print("Age  :", self.age)

class Employee(Person):
    def __init__(self, name, age, salary):
        # calling parent constructor
        super().__init__(name, age)
        self.salary = salary


    def show_salary(self):
        print("Salary :", self.salary)

emp1 = Employee("John", 25, 50000)
emp1.show_details()
emp1.show_salary()
```

```python
# using static method and class method 
class Employee:
    company = "ABC Technologies"
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    # instance method
    def show_employee(self):
        print("Name    :", self.name)
        print("Salary  :", self.salary)
        print("Company :", Employee.company)

    # class method
    @classmethod
    def change_company(cls, new_company):
        cls.company = new_company

    # static method
    @staticmethod
    def is_valid_salary(salary):
        return salary > 0

emp1 = Employee("John", 50000)
emp2 = Employee("Mary", 60000)

print("Before Company Change")
print("---------------------")
emp1.show_employee()
print()
emp2.show_employee()


print()
print("Using Class Method")
print("------------------")

Employee.change_company("XYZ Solutions")
emp1.show_employee()
print()
emp2.show_employee()


print()
print("Using Static Method")
print("-------------------")
print(Employee.is_valid_salary(5000))
print(Employee.is_valid_salary(-100))
```

```python
# Inheritance and Duck-typing.
# Parent Class
class Animal:
    def sound(self):
        print("Animal makes sound")

# Child Classes
class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

# Duck Typing Class
class Robot:
    def sound(self):
        print("Robot speaks digitally")

# common function
def make_sound(obj):
    obj.sound()

dog = Dog()
cat = Cat()
robot = Robot()

make_sound(dog)
make_sound(cat)
make_sound(robot)
```

```python
### OO Design: Pythonisms and best practices
class Employee:
    # class variable
    company = "ABC Technologies"

    def __init__(self, emp_id, name, salary):
        # encapsulation
        self._emp_id = emp_id
        self._name = name
        self._salary = salary


    # string representation
    def __repr__(self):
        return (
            f"Employee("
            f"id={self._emp_id}, "
            f"name='{self._name}', "
            f"salary={self._salary})"
        )


    # property getter
    @property
    def salary(self):
        return self._salary

    # property setter with validation
    @salary.setter
    def salary(self, value):
        if value < 0:
            raise ValueError("Salary cannot be negative")

        self._salary = value


    # instance method
    def show_details(self):
        print(
            f"ID: {self._emp_id}, "
            f"Name: {self._name}, "
            f"Salary: {self._salary}"
        )


    # class method
    @classmethod
    def change_company(cls, new_company):
        cls.company = new_company


    # static method
    @staticmethod
    def is_valid_age(age):
        return age >= 18

emp = Employee(101, "John", 50000)
print(emp)
print()
emp.show_details()
print()
print("Salary :", emp.salary)
emp.salary = 70000
print("Updated Salary :", emp.salary)
print()
print(Employee.is_valid_age(25))
print()
Employee.change_company("XYZ Solutions")
print(Employee.company)
```

```python
##  Creating and using decorators in Python.
## Creating decorators with arguments (parameterized decorators).

# decorator function
def log_function_call(func):
    def wrapper(*args, **kargs):
        print("Function execution started")
        func(*args, **kargs)
        print("Function execution completed")
    return wrapper

# applying decorator
@log_function_call
def display_message(name, age):
    print("Welcome to Python Decorators")
    print(f'name:{name}')
    print(f'name:{age}')

display_message('Mahesh', 47)
```

```python
##  Use-case examples and best practices for implementing decorators
import time

# decorator
def execution_time(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()

        print(f"Execution Time: {end - start:.4f} seconds")
        return result
    return wrapper


# use-case:
# monitoring performance of heavy functions
@execution_time
def process_large_data():
    total = 0
    for num in range(1, 1000000):
        total += num

    print("Processing Completed")

process_large_data()
```

```python
## Decorating classes and methods of a class

# decorator for methods
def log_method(func):
    def wrapper(self):
        print("Method execution started")
        func(self)
        print("Method execution completed")
    return wrapper

# decorator for classes
def log_class(cls):
    print(f"Class '{cls.__name__}' is created")
    return cls

@log_class
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    @log_method
    def show_details(self):
        print("Name   :", self.name)
        print("Salary :", self.salary)



emp = Employee("John", 50000)
print()
emp.show_details()
```
