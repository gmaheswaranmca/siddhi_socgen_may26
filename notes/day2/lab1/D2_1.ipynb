```python
# what is function? function is named code, takes arg as input, return value as output 
# find max of 3 nums
# define function 
# call by value
def find_max_of_3(num1, num2, num3):
    max_num = num1 
    if num2 > max_num:
        max_num = num2
    #
    if num3 > max_num:
        max_num = num3
    #
    return max_num 

# positional args
print(find_max_of_3(10, 20, 30))
print(find_max_of_3(5.2, 7.4, 3.1))
print(find_max_of_3((8,10), (4,20), (2,50)))

# keyword args
print(find_max_of_3(num1=10, num2=20, num3=30)) 
print(find_max_of_3(num3=10, num1=20, num2=30)) 
```

```python
# Defult args
# call by value 
def find_net_salary(salary, hra_per = 0, da_per = 0):
    net_salary = salary

    if hra_per == 0 and da_per == 0:
        return net_salary
    
    if hra_per > 0:
        # process to calc hra, net_salary
        hra = (hra_per / 100) * salary 
        net_salary += hra
    #

    if da_per > 0:
        # process to calc hra, net_salary
        da = (da_per / 100) * salary 
        net_salary += da
    # 

    return net_salary 
# 


print(find_net_salary(300000))
print(find_net_salary(300000, 20))

print(find_net_salary(300000, da_per = 10))

print(find_net_salary(300000, 20, 10))

print(find_net_salary(300000, da_per = 20, hra_per = 10))

print(find_net_salary(salary = 300000, da_per = 20, hra_per = 10))
```

```python
# Defult args
# call by value - return multiple values
def find_net_salary(salary, hra_per = 0, da_per = 0):
    net_salary = salary

    if hra_per == 0 and da_per == 0:
        return 0, 0, net_salary
    
    hra = 0
    if hra_per > 0:
        # process to calc hra, net_salary
        hra = (hra_per / 100) * salary 
        net_salary += hra
    #

    da = 0
    if da_per > 0:
        # process to calc hra, net_salary
        da = (da_per / 100) * salary 
        net_salary += da
    # 

    return hra, da, net_salary 
# 


print(find_net_salary(300000))
print(find_net_salary(300000, 20))
print(find_net_salary(300000, 20, 10))
```

```python
# count_prime for given sequence (via return statement in is_prime fn)
import math 

# call by value 
def is_prime(num): 
    for mid in range(2, int(math.sqrt(num)) + 1): 
        if num % mid == 0:
            return False 
        #
    #

    return True 
#

# call by reference 
def count_prime(nums):
    count = 0 
    for num in nums:
        if is_prime(num):
            count += 1
        #
    #
    return count 
#

print(count_prime([10, 11, 31, 50, 51, 61, 70, 79, 80]))
print(count_prime((10, 11, 31, 50, 51, 61, 70, 79, 80)))
```

```python
# count_prime for given sequence - v2 (via break statement in is_prime fn)
import math 

# call by value 
def is_prime(num):
    result = True
    for mid in range(2, int(math.sqrt(num)) + 1):
        if num % mid == 0:
            result = False 
            break  
        #
    #

    return result 
#
# call by reference 
def count_prime(nums):
    count = 0 
    for num in nums:
        if is_prime(num):
            count += 1
        #
    #
    return count 
#

print(count_prime([10, 11, 31, 50, 51, 61, 70, 79, 80]))
print(count_prime((10, 11, 31, 50, 51, 61, 70, 79, 80)))
```

```python
# find occurance freq
# call by reference 
def find_occur_freq(nums):
    freqs = dict()
    for num in nums:
        freqs[num] = freqs.get(num, 0) + 1
    #
    return freqs
#

ages = [1, 3, 1, 3, 5, 2, 3, 20, 10, 5, 3, 1, 20]
freqs = find_occur_freq(ages)
print(freqs) # dict == key-value (hashing)

# remove duplicate ages
ages_no_dup = set(ages) # set == keys (hashing)
print(ages_no_dup) 
```

```python
# recursive function 
# n! = n * (n-1)!  -> 5! = 5 * 4! -> 5! = 5 * 4 * 3! -> 5 * 4 * 3 * 2! -> 5 * 4 * 3 * 2 * 1! 
#                  <- 5 * 4 * 3 * 2 * 1 <- 5 * 4 * 3 * 2 <- 5 * 4 * 6 <- 5 * 24 <- 120
def find_fact(num):
    if num <= 1: # base cond
        return 1
    #
    return num * find_fact(num-1) # fn calls itself
#

print(find_fact(5))
print(find_fact(4))
print(find_fact(6))
```

```python
# var arg fn 

# find max of n nums
# define function 
# call by value
def find_max_of_n(num1, num2, num3 = None, *others):
    max_num = num1 
    if num2 > max_num: # max of 2
        max_num = num2
    #
    
    if num3 == None:
        return max_num 
    #

    if num3 > max_num: # max of 3
        max_num = num3
    #

    if len(others) == 0: 
        return max_num
    #

    for other_num in others: # max of n
        if other_num > max_num:
            max_num = other_num
        #
    #
    return max_num 
#

# positional args
print(find_max_of_n(10, 20))
print(find_max_of_n(10, 20, 30))
print(find_max_of_n(5.2, 7.4, 3.1))
print(find_max_of_n((8,10), (4,20), (2,50)))

# keyword args
print(find_max_of_n(num1=10, num2=20, num3=30)) 
print(find_max_of_n(num3=10, num1=20, num2=30)) 

# variable args 
print(find_max_of_n(10, 20, 40, 30))
print(find_max_of_n(10, 20, 40, 30, 70, 25))
```

```python
# var keyword arg fn 

# find max of n nums
# define function 
# call by value
def find_max_of_n(num1, num2, num3 = None, **others):
    max_num = num1 
    if num2 > max_num: # max of 2
        max_num = num2
    #
    
    if num3 == None:
        return max_num 
    #

    if num3 > max_num: # max of 3
        max_num = num3
    #

    if len(others) == 0: 
        return max_num
    #

    for other_key in others: # max of n
        other_num = others[other_key]
        if other_num > max_num:
            max_num = other_num
        #
    #
    return max_num 
#

# keyword variable args 
print(find_max_of_n(num1 = 10, num2 = 20, num3 = 40, num4 = 30))
print(find_max_of_n(num1 = 10, num2 = 20, num3 = 40, num4 = 30, num5 = 70, num6 = 25))
```

```python
# A lambda function is an anonymous small function.
# syntax : lambda arguments : expression
two_times = lambda a : a * 2
print(two_times(5))
print(two_times(90))

add_2 = lambda a,b : a + b
print(add_2(5, 6))
print(add_2(90, 100))

nums = [10, 21, 30]
print(list(map(two_times, nums)) , nums)

print(list(map(lambda num:num+3, nums)), nums)

print(list(filter(lambda num:num%2==0, nums)), nums)
```
