```python
# multiple exceptions and generic exception 
# raise exception
try:
    a = int(input("Enter number:"))
    
    # This 
    # if a == 0:
    #    raise ZeroDivisionError("Age cannot be negative")
    # Or 
    b = 10 / a

    print(b)

except ZeroDivisionError:
    print("Division by zero")

except ValueError:
    print("Invalid input")

except Exception as e:
    print("Error:", e)

else: # if exception is not there
    print("Division successful")
    
finally: # for both exception and no exc 
    print("Program completed")
```

```python
# Custome Exception and raising exception
class InvalidSalary(Exception):
    pass

salary = -1000

if salary < 0:
    raise InvalidSalary("Salary cannot be negative")
```

```python
# Custome Exception and raising exception
# Handle exception
class InvalidAge(Exception):
    pass

try:

    age = -1

    if age < 0:
        raise InvalidAge("Invalid age")

except InvalidAge as e:
    print(e)
```

```python
# File Handling Error 

f = None

try:
    f = open("sample.txt", "r")
    print(f.read())

except FileNotFoundError:
    print("File not found")

finally:

    if f:
        f.close()

    print("File closed")
```
