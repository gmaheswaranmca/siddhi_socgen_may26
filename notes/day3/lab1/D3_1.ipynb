```python
# write ages into text file
ages = [1,5,6,10,1,5,8,9,10,1] 
ages_tuple = (1,5,6,10,1,5,8,9,10,1) 
ages_freq = {1: 3, 5 : 2, 6: 1, 10:2, 8:1, 9:1} 
ages_distict = {1, 5, 6, 10, 8, 9}

FILE_NAME = 'ages.txt'
with open(FILE_NAME, 'w') as outfile:
    outfile.write(str(ages) + '\n')
    outfile.write(str(ages_tuple) + '\n')
    outfile.write(str(ages_freq) + '\n')    
    outfile.write(str(ages_distict))
print(f'Ages written into file {FILE_NAME}')
```

```python
# read ages from text file
import ast
FILE_NAME = 'ages.txt'
with open(FILE_NAME, 'r') as infile:
    ages_list_str = infile.readline()
    ages_tuple_str = infile.readline()
    ages_freq_str = infile.readline()    
    ages_distict_str = infile.readline()

    print(type(ages_list_str))
    print(ages_tuple_str)
    print(ages_freq_str)
    print(ages_distict_str)

    ages_list = ast.literal_eval(ages_list_str)
    ages_tuple = ast.literal_eval(ages_tuple_str)
    ages_freq = ast.literal_eval(ages_freq_str)
    ages_distict = ast.literal_eval(ages_distict_str)

    print(ages_list)
    print(ages_tuple)
    print(ages_freq)
    print(ages_distict)
```
