```python
### repo - mem - list[dict]
employees = []

def new_id():
    if len(employees) == 0:
        return 1
    
    return employees[len(employees)-1]['id'] + 1

def create_emp(emp):
    emp['id'] = new_id()
    employees.append(emp)

def read_all_employees():
    return employees

def read_emp_by_id(id):
    for emp in employees:
        if emp['id'] == id:
            return emp
        #
    #
    return None

def update_emp_by_id(id, new_age):
    for emp in employees:
        if emp['id'] == id:
            emp['age'] = new_age
        #
    #

def delete_emp_by_id(id):
    for index, emp in enumerate(employees):
        if emp['id'] == id:
            employees.pop(index)
            return
        #
    #
```

```python
# test read_all
print(read_all_employees())
```

```python
### test create
create_emp({'name':'Mahesh', 'age':47})
create_emp({'name':'Aparna', 'age':27})
create_emp({'name':'Aniruth', 'age':23})

print(read_all_employees())
```

```python
### test read_by_id
id = 2
print(read_emp_by_id(2))
```

```python
### test update_by_id
id = 3
new_age = 30
update_emp_by_id(id, new_age)
print(read_all_employees())
```

```python
### test delete_by_id
delete_emp_by_id(1)
print(read_all_employees())
```
