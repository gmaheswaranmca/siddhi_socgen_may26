```python
### repo - sqlite (persistent store) - list[dict]
import sqlite3

DB_NAME = 'emp.db'


def get_connection():
    return sqlite3.connect(DB_NAME)


def create_table():

    conn = get_connection()

    cursor = conn.cursor()

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS employee(
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT,
            age INTEGER
        )
    """)

    conn.commit()
    conn.close()


create_table()


### CREATE
def create_emp(emp):

    conn = get_connection()

    cursor = conn.cursor()

    cursor.execute(
        "INSERT INTO employee(name, age) VALUES(?, ?)",
        (emp['name'], emp['age'])
    )

    conn.commit()
    conn.close()


### READ ALL
def read_all_employees():

    conn = get_connection()

    cursor = conn.cursor()

    cursor.execute("SELECT id, name, age FROM employee")

    rows = cursor.fetchall()

    conn.close()

    employees = []

    for row in rows:
        employees.append({
            'id': row[0],
            'name': row[1],
            'age': row[2]
        })

    return employees


### READ BY ID
def read_emp_by_id(id):

    conn = get_connection()

    cursor = conn.cursor()

    cursor.execute(
        "SELECT id, name, age FROM employee WHERE id = ?",
        (id,)
    )

    row = cursor.fetchone()

    conn.close()

    if row is None:
        return None

    return {
        'id': row[0],
        'name': row[1],
        'age': row[2]
    }


### UPDATE
def update_emp_by_id(id, new_age):

    conn = get_connection()

    cursor = conn.cursor()

    cursor.execute(
        "UPDATE employee SET age = ? WHERE id = ?",
        (new_age, id)
    )

    conn.commit()
    conn.close()


### DELETE
def delete_emp_by_id(id):

    conn = get_connection()

    cursor = conn.cursor()

    cursor.execute(
        "DELETE FROM employee WHERE id = ?",
        (id,)
    )

    conn.commit()
    conn.close()
```

```python
### test read_all
print(read_all_employees())
```

```python
### test create
create_emp({'name':'Mahesh', 'age':47})
create_emp({'name':'Aparna', 'age':27})
create_emp({'name':'Aniruth', 'age':23})

print(read_all_employees())
```

```python
### test read_by_id
id = 2
print(read_emp_by_id(2))
```

```python
### test update_by_id
id = 3
new_age = 30
update_emp_by_id(id, new_age)
print(read_all_employees())
```

```python
### test delete_by_id
delete_emp_by_id(1)
print(read_all_employees())
```
